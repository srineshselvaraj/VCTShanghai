### Scraping Data from vlr.gg

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_url(team):
    if team == "100T":
        url = 'https://www.vlr.gg/team/stats/120/100-thieves/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "G2":
        url = 'https://www.vlr.gg/team/stats/11058/g2-esports/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "LEV":
        url = 'https://www.vlr.gg/team/stats/2359/leviat-n/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "FNC":
        url = 'https://www.vlr.gg/team/stats/2593/fnatic/?event_id=all&date_start=2024-02-16&date_end='
    elif team == "TH":
        url = 'https://www.vlr.gg/team/stats/1001/team-heretics/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "FUT":
        url = 'https://www.vlr.gg/team/stats/1184/fut-esports/?event_id=all&date_start=2024-02-16&date_end='
    elif team == "PRX":
        url = 'https://www.vlr.gg/team/stats/624/paper-rex/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "GEN":
        url = 'https://www.vlr.gg/team/stats/17/gen-g/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "T1":
        url = 'https://www.vlr.gg/team/stats/14/t1/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "EDG":
        url = 'https://www.vlr.gg/team/stats/1120/edward-gaming/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    elif team == "FPX":
        url = 'https://www.vlr.gg/team/stats/11328/funplus-phoenix/?event_id=all&date_start=2024-02-16&date_end='
    elif team == "DRG":
        url = 'https://www.vlr.gg/team/stats/11981/dragon-ranger-gaming/?event_id=all&core_id=all&date_start=2024-02-16&date_end='
    else:
        url = ''
    return url
```

```python
def get_data(team):
    url = get_url(team)
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    table = soup.find('table', class_ ="wf-table mod-team-maps")
    stats = table.find_all('tr', class_ = None)

    columns = ["map", "times_played", "win_rate", "wins", "losses", "atk_1st", "def_1st", 
               "atk_win_rate", "atk_wins", "atk_losses", "def_win_rate", "def_wins", "def_losses"]
    list = []

    rows = []
    for stat in stats:
        rows.append(stat.get_text().replace("\t", " ").replace("\n",""))

    for row in rows:
        data = row.split(" ")
        data = [i for i in data if i != '']
        if(data[0] == "Map"):
            continue
        if data[1] != "(0)":
            data = [data[0], int(data[1][1:2]), data[2], int(data[3]), int(data[4]), int(data[5]), int(data[6]), data[7], 
                    int(data[8]), int(data[9]), data[10], int(data[11]), int(data[12])]
            list.append(data)

    df = pd.DataFrame(list, columns=columns)
    return df
```

```python
df_100t = get_data("100T")
df_g2 = get_data("G2")
df_lev = get_data("LEV")
df_fnc = get_data("FNC")
df_th = get_data("TH")
df_fut = get_data("FUT")
df_prx = get_data("PRX")
df_gen = get_data("GEN")
df_t1 = get_data("T1")
df_edg = get_data("EDG")
df_fpx = get_data("FPX")
df_drg = get_data("DRG")

vct_data = {"100T":df_100t, "G2":df_g2, "LEV":df_lev, "FNC":df_fnc, "TH":df_th, "FUT":df_fut, 
            "PRX":df_prx, "GEN":df_gen, "T1":df_t1, "EDG":df_edg, "FPX":df_fpx, "DRG":df_drg}
```
